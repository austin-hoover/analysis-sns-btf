```python
data = file['/scandata']
for item in data.dtype.fields.items():
    print(item)
```

```python
magnets = ['DCV01', 'DCV02']
for magnet in magnets:
    fig, ax = pplt.subplots(figsize=(8, 2))
    kws = dict(color='black', marker='.', ms=3)
    ax.plot(data[magnet + '_CurrentRb'], **kws)
    ax.plot(data[magnet + '_CurrentSet'], alpha=0.2, **kws)
    ax.format(ylabel=magnet, xlabel='Step')
    plt.show()
```

```python
cut = 800
steps = [20, 20]
n_avg = 3
ys = []
yraws = []
idx_ = np.arange(0, len(data), n_avg)
for i in tqdm(idx_):
    y = np.mean(data['BPM04_4_Y'][i:i+3, :cut], axis=0)
    u = np.mean(data['BPM04_4_MAG_U'][i:i+3, :cut], axis=0)
    d = np.mean(data['BPM04_4_MAG_D'][i:i+3, :cut], axis=0)
    yraw = 8.0 * np.log10(u / d)
    ys.append(y)
    yraws.append(yraw)
ys = np.array(ys)
yraws = np.array(yraws)
```

```python
lo = 562
hi = 592
xvals = np.arange(len(ys[0]))

for i in range(10):
    y = ys[i]
    yraw = yraws[i]
    fig, ax = pplt.subplots(figsize=(5.0, 2.0))
    ax.plot(xvals, y, color='black', marker='.', ms=3)
    ax.plot(xvals, yraw, color='grey', marker='.', ms=3)
    ax.axvspan(lo, hi, color='black', alpha=0.08, ec='None')
    pad = int(0.05 * max(lo, hi))
    ax.format(xlim=(lo - pad, hi + pad))
```

```python
_ys = yraws
```

```python
yavgs = []
for i, y in enumerate(_ys):
    yavg = np.mean(y[lo:hi])
    yavgs.append(yavg)
yavgs = np.array(yavgs)

bs = np.abs(data[idx_, 'BS34'])
```
