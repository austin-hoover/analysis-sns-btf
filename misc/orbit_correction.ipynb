```python
datadir = './Diagnostics/Data/Measurements/2022-05-09/'
filenames = os.listdir(datadir)
filenames
```

```python
filename = '220509085306-orbit_correction.h5'
```

```python
file = h5py.File(join(datadir, filename), 'r')
```

```python
# Errors and warnings from log
for i in range(file['log'].size):
    if not(file['/log'][i, 'level'] == 'INFO'.encode('utf')):
        timestr = datetime.fromtimestamp(file['/log'][0, 'timestamp']).strftime("%m/%d/%Y, %H:%M:%S")
        print(f"{timestr} {file['log'][i, 'message']}")

# Configuration data
for key in file['/config'].keys():
    print(f"{key}")
    print("--------------")
    for name in file['/config'][key].dtype.names:
        print(f"{name}: {file['config'][key][name]}")
    print()
```

```python
data = file['/scandata']
for item in data.dtype.fields.items():
    print(item)
```

```python
magnets = ['DCV01', 'DCV02']
for magnet in magnets:
    fig, ax = pplt.subplots(figsize=(8, 2))
    kws = dict(color='black', marker='.', ms=3)
    ax.plot(data[magnet + '_CurrentRb'], **kws)
    ax.plot(data[magnet + '_CurrentSet'], alpha=0.2, **kws)
    ax.format(ylabel=magnet, xlabel='Step')
    plt.show()
```

```python
cut = 800
steps = [20, 20]
n_avg = 3
ys = []
yraws = []
idx_ = np.arange(0, len(data), n_avg)
for i in tqdm(idx_):
    y = np.mean(data['BPM04_4_Y'][i:i+3, :cut], axis=0)
    u = np.mean(data['BPM04_4_MAG_U'][i:i+3, :cut], axis=0)
    d = np.mean(data['BPM04_4_MAG_D'][i:i+3, :cut], axis=0)
    yraw = 8.0 * np.log10(u / d)
    ys.append(y)
    yraws.append(yraw)
ys = np.array(ys)
yraws = np.array(yraws)
```

```python
lo = 562
hi = 592
xvals = np.arange(len(ys[0]))

for i in range(10):
    y = ys[i]
    yraw = yraws[i]
    fig, ax = pplt.subplots(figsize=(5.0, 2.0))
    ax.plot(xvals, y, color='black', marker='.', ms=3)
    ax.plot(xvals, yraw, color='grey', marker='.', ms=3)
    ax.axvspan(lo, hi, color='black', alpha=0.08, ec='None')
    pad = int(0.05 * max(lo, hi))
    ax.format(xlim=(lo - pad, hi + pad))
```

```python
_ys = yraws
# _ys = ys
```

```python
yavgs = []
for i, y in enumerate(_ys):
    yavg = np.mean(y[lo:hi])
    yavgs.append(yavg)
yavgs = np.array(yavgs)

# signal = np.abs(data[idx_, 'BS34'])
signal_pv = 'Cam06_Integral'
signal = np.abs(data[idx_, signal_pv])
```

```python
from scipy.signal import find_peaks

idx_peak = find_peaks(-np.abs(yavgs))[0]

fig, ax = pplt.subplots(
    figsize=(4.0, 2.0)
)
ax.plot(yavgs, color='black')
ax.plot(idx_peak, yavgs[idx_peak], lw=0, marker='.', color='red', label='ybpm')
ax.format(ylabel='<y BPM> [mm]', xlabel='Step')
ax2 = ax.twinx()
ax2.plot(idx_peak, signal[idx_peak], lw=0, marker='.', color='blue', label='cam integral')
ax2.format(
    color='gray', 
    ylabel=signal_pv,
)
ax2.plot(signal, color='black', alpha=0.1)
ax.legend(loc='top')
plt.show()
```

```python
for ipeak in idx_peak:
    m1 = data[idx_[ipeak], magnets[0] + '_CurrentRb']
    m2 = data[idx_[ipeak], magnets[1] + '_CurrentRb']
    print(f'{ipeak}, {yavgs[ipeak]:.2f}, {signal[ipeak]:.2f}, {m1:.2f}, {m2:.2f}')
```

```python
fig, ax = pplt.subplots(
    figsize=(4.0, 1.75)
)
ax.plot(yavgs, color='black')
ax.format(ylabel='<y BPM> [mm]', xlabel='Step')
ax2 = ax.twinx()
ax2.format(
    color='gray', 
    ylabel=signal_pv,
#     ylabel='Beam current [-mA]'
)
ax2.plot(signal, color='black', alpha=0.1)
plt.savefig('_output/results1da')
```
